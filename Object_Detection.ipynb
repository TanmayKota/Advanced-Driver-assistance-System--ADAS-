```python
# All required imports
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
input_image_3 = cv2.imread(".//data//Task_03.png")                                          # Accesing Input Image 3    
input_image_3_color_correction=cv2.cvtColor(input_image_3,cv2.COLOR_BGR2RGB)                # Converting Input Image color scheme from BGR to RGB because matplotlib.pyplot can display images only in RGB format
plt.title('Original Image')
plt.axis('off')
plt.imshow(input_image_3_color_correction)                                                  # Displaying the input image
plt.show()

input_template = cv2.imread(".//data//Template.png")                                        # Accesing Template
input_template_color_correction=cv2.cvtColor(input_template,cv2.COLOR_BGR2RGB)              # Converting Input Image color scheme from BGR to RGB because matplotlib.pyplot can display images only in RGB format
plt.title('Template')
plt.axis('off')
plt.imshow(input_template_color_correction)                                                 # Displaying the Template
plt.show()

# Matching the template to the image
h,w=input_template.shape[0],input_template.shape[1]                                         # Accessing the height and width of the Template
result=cv2.matchTemplate(input_image_3,input_template,cv2.TM_CCOEFF_NORMED)                 # Matching the template to the input image
minVal,maxVal,minLoc,maxLoc=cv2.minMaxLoc(result)                                           # Accessing the minimum and maximum Location and Variable
top_left=maxLoc                                                                             # Assigning maximum location to top left because the function TM_CCOEFF_NORMED demands it
bottom_right=(top_left[0]+w,top_left[1]+h)                                                  # Calculating bottom right to get the rectangle on the result image
rectangle=cv2.rectangle(input_image_3,top_left,bottom_right,(0,255,255),2)                  # Creating the rectangle on the image
result=cv2.cvtColor(rectangle,cv2.COLOR_BGR2RGB)                                            # Converting Result Image color scheme from BGR to RGB because matplotlib.pyplot can display images only in RGB format
plt.title('Template matched Image')
plt.axis('off')
plt.imshow(result)                                                                          # Displaying the final result with the template matched
plt.show()
```
